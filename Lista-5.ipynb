```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import gamma
α=3
λ=2
```

```python
steps=np.arange(10,1001,10)
result=[]
yLam=np.array([])
yAlf=np.array([])
for i in steps:
    lam=np.array([])
    alf=np.array([])
    for _ in range(1000):
        X=np.random.gamma(α,1/λ,i)
        estLam=np.sum(X)/(np.sum(X**2)-np.sum(X)**2/i)
        estAlf=np.sum(X)**2/(i*(np.sum(X**2)-np.sum(X)**2/i))
        lam=np.append(lam,estLam)
        alf=np.append(alf,estAlf)
    yLam=np.append(yLam,np.mean(lam))
    yAlf=np.append(yAlf,np.mean(alf))
    
plt.plot(steps,yLam, label='Lambda')
plt.hlines(y=λ,xmin=0,xmax=1000)
plt.plot(steps,yAlf, label='Alfa')
plt.hlines(y=α,xmin=0,xmax=1000)
plt.legend()
plt.show()

plt.plot(steps,yLam-λ, label='Lambda')
plt.plot(steps,yAlf-α, label='Alfa')
plt.legend()
plt.show()
```

```python
import numpy as np
import scipy.stats as st

α=3
λ=2


step=range(10,1000,2)
mean_lambd=np.array([])
mean_alpha=np.array([])

for n in step:
    lambd_results=np.array([])
    alpha_results=np.array([])
    for _ in range(1000):
        x=np.random.gamma(α, 1/λ, n)
        lambd=np.sum(x)/(np.sum(x**2)-np.sum(x)**2/n) #estymator lambdy który wyliczyliśmy
        alpha=(np.sum(x))**2/(n*(np.sum(x**2)-1/n*(np.sum(x)**2))) #estymator alphy który wyliczyliśmy
        
        lambd_results=np.append(lambd_results,lambd)
        alpha_results=np.append(alpha_results,alpha)
        
    mean_lambd=np.append(mean_lambd,np.mean(lambd_results))
    mean_alpha=np.append(mean_alpha,np.mean(alpha_results))
    
plt.plot(step,mean_lambd,label="estymator lambdy", color="green")
plt.hlines(y=λ,xmin=0,xmax=1000)
plt.legend()
plt.show()
plt.plot(step,mean_alpha,label="estymator alphy", color="hotpink")
plt.hlines(y=α,xmin=0,xmax=1000)
plt.legend()
plt.show()
    
```

```python
import numpy as np
import matplotlib.pyplot as plt


def gen(a,n):
    u = np.random.rand(n)
    return (u**(1/(a+1)))

a=2
n=1000

a1=[]
a2=[]
for _ in range(500):
    X=gen(a,n)
    a1.append((1/(1-1/n*np.sum(X)))-2)
    a2.append(-n/np.sum(np.log(X))-1)
    
plt.boxplot(a1)
plt.show()
plt.boxplot(a2)
plt.show()
```
