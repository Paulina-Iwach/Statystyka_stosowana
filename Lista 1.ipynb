```python
lambd=1
alpha=1
theta=3
n=1000
u=np.random.rand(n)
x=((lambd/((1-u)**(1/alpha)))-lambd)**(1/theta)
x
plt.hist(x,density=True, bins=[0.25*i for i in range(0,30)])
```

```python
x_f=np.linspace(0, 10, num=10**4)
y_f=(theta*lambd*x_f**(theta-1)*lambd**alpha)/((lambd+x_f**theta)**(alpha+1))

plt.hist(x,density=True, bins=[0.25*i for i in range(0,30)], color="pink")
plt.plot(x_f, y_f,color="red")
```

```python
lambd=1
alpha=1
theta=3
n=1000
u=np.random.rand(n)
x=((lambd/((1-u)**(1/alpha)))-lambd)**(1/theta)

x_F=np.linspace(0, 6.0, num=10**3)
y_F=1-((lambd/(lambd+x_F**theta))**alpha)
   
    
args=np.sort(x)
y=np.arange(len(args))/float(len(args))
    
    
plt.plot(x_F,y_F,color="blue")
plt.plot(args,y, color="orange")
plt.xlim(0,10)
plt.show()
```
