```python
alpha=0.02
q=1-(alpha/2)
M=1000
n=[20,50,100,1000]


results=[]

for n in n:
    count=0
    for _ in range(M):
        x=st.norm.rvs(size=n,loc=,scale=)
        mean_x=np.mean(x)
        a=mean_x-st.norm.ppf(q,loc=0,scale=1)*/np.sqrt(n)
        b=mean_x+st.norm.ppf(q,loc=0,scale=1)*/np.sqrt(n)

        if a<=<=b:
            count+=1
    results.append(count/M)

n=[20,50,100,1000]
for i in range(len(results)):
    #print("dla n="+str(n[i])+" dokadno przedziau ufnoci wynosi "+str(results[i]))
    print("dla n={} dokadno przedziau ufnoci wynosi {}.".format(n[i],results[i]))
```
